```python
import pandas as pd
import matplotlib.pyplot as plt


# Replace 'your_column_names' with the actual column names from your dataset
column_names = ['DELIV_STORE_NAME', 'DELIV_MISSING_INCORRECT_REPORT', 'WAS_REQUESTED', 'WAS_MISSING', 'WAS_SUBBED', 'WAS_FOUND', 'WAS_FOUND', 'ITEM_NAME', 'ITEM_CATEGORY', 'SUBSTITUTE_ITEM_NAME', 'SUBSTITUTE_ITEM_CATEGORY']

# Iterate over each column and plot the column count
for column_name in column_names:
    value_counts = df[column_name].value_counts()

    # Plot the column count
    plt.figure(figsize=(10, 6))
    value_counts.plot(kind='bar')
    plt.title(f'Column Count for {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Count')

plt.show()
```
